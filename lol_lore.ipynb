```python
# List of champions
champions = [
    "aatrox", "ahri", "akali", "akshan", "alistar", "amumu", "anivia", "annie", "aphelios", "ashe",
    "aurelionsol", "azir", "bard", "belveth", "blitzcrank", "brand", "braum", "caitlyn", "camille",
    "cassiopeia", "chogath", "corki", "darius", "diana", "drmundo", "draven", "ekko", "elise",
    "evelynn", "ezreal", "fiddlesticks", "fiora", "fizz", "galio", "gangplank", "garen", "gnar",
    "gragas", "graves", "gwen", "hecarim", "heimerdinger", "illaoi", "irelia", "ivern", "janna",
    "jarvaniv", "jax", "jayce", "jhin", "jinx", "kaisa", "kalista", "karma", "karthus", "kassadin",
    "katarina", "kayle", "kayn", "kennen", "khazix", "kindred", "kled", "kogmaw", "leblanc", "leesin",
    "leona", "lillia", "lissandra", "lucian", "lulu", "lux", "malphite", "malzahar", "maokai",
    "masteryi", "milio", "missfortune", "mordekaiser", "morgana", "naafiri", "nami", "nasus",
    "nautilus", "neeko", "nidalee", "nilah", "nocturne", "nunu", "olaf", "orianna", "ornn",
    "pantheon", "poppy", "pyke", "qiyana", "quinn", "rakan", "rammus", "reksai", "rell", "renataglasc",
    "renekton", "rengar", "riven", "rumble", "ryze", "samira", "sejuani", "senna", "seraphine", "sett",
    "shaco", "shen", "shyvana", "singed", "sion", "sivir", "skarner", "sona", "soraka", "swain",
    "sylas", "syndra", "tahmkench", "taliyah", "talon", "taric", "teemo", "thresh", "tristana",
    "trundle", "tryndamere", "twistedfate", "twitch", "udyr", "urgot", "varus", "vayne", "veigar",
    "velkoz", "vex", "vi", "viego", "viktor", "vladimir", "volibear", "warwick", "monkeyking", "xayah",
    "xerath", "xinzhao", "yasuo", "yone", "yorick", "yuumi", "zac", "zed", "ziggs", "zilean", "zoe", "zyra"
]

print(f"The total number of champions: {len(champions)}")

# Base URL for the champion story in Korean
base_url = "https://universe.leagueoflegends.com/ko_KR/story/champion/"

# Function to scrape the Korean name and background story of a champion
def scrape_champion_data(champion):
    url = base_url + champion + "/"
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract the Korean name from the <title> tag
        korean_name = soup.find('title').text.split('-')[0].strip()

        # Extract the background story from the meta description
        meta_description = soup.find('meta', {'name': 'description'})
        if meta_description:
            background_story = meta_description.get('content').replace('\n', ' ').strip()
        else:
            background_story = "No background story available"

        return korean_name, background_story
    else:
        return None, None

# Open the CSV file for writing
with open("champion_bs.csv", "w", newline='', encoding='utf-8') as csvfile:
    # Define the column headers
    fieldnames = ['url-name', 'korean-name', 'background-story']

    # Create a CSV writer object
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # Write the header
    writer.writeheader()

    # Scrape data for each champion and write to CSV
    for champion in champions:
        korean_name, background_story = scrape_champion_data(champion)
        if korean_name and background_story:
            writer.writerow({
                'url-name': champion,
                'korean-name': korean_name,
                'background-story': background_story
            })
            print(f"Scraped data for {champion}: {korean_name}")
        else:
            print(f"Failed to scrape data for {champion}")

print("Data scraping complete. Saved to champion_bs.csv")
```

```python
client = OpenAI(
    api_key=userdata.get('OPENAI_TOKEN')
)

# Define the base prompt to append the background story
base_prompt = """
다음 텍스트는 온라인 게임 리그 오브 레전드의 한 캐릭터의 배경 설명이다.
다음 텍스트에서 질문과 답을 생성하세요.
각 문단에서 핵심 정보나 중요한 사건, 캐릭터의 행동 또는 결정에 관한 질문을 만드세요.
질문은 짧고 명확하게 하며, 답은 구체적이고 텍스트에서 직접적으로 추론할 수 있도록 작성하세요.
다음과 같은 형식으로만 질문과 답을 구성하세요.

"대부분의 필멸자가 알고 있는 현실 차원은 무엇인가?", "대부분의 필멸자는 물질 세계라는 하나의 현실 차원만 알고 있다."
"오로라가 유년 시절을 보낸 곳은 어디인가?", "오로라는 브뤼니 부족의 고향이자 외딴 마을인 아무우에서 유년 시절을 보냈다."
"오로라가 자신을 이해해준 유일한 가족 구성원은 누구인가?", "오로라의 이모할머니 하부우가 오로라를 진심으로 받아들였다."
"오로라가 유년 시절에 보고 경험한 세계는 어떤 세계였나?", "오로라는 필멸 세계와 영혼 세계의 장막을 꿰뚫는 특별한 세계를 경험했다."
"오로라는 무엇을 연구하며 기록했는가?", "오로라는 아무우의 영혼과 뒤얽힌 세계를 연구하며 세심하게 기록했다."
"""

# Function to interact with OpenAI's GPT-4 model and get responses
def generate_questions_and_answers(story):
    # Combine the base prompt with the story
    prompt = base_prompt + "\n\n" + story

    try:
        # Send the prompt to GPT-4 and get the response
        chat_completion = client.chat.completions.create(
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            model="gpt-4o-mini",
        )

        # Extract the generated text (questions and answers)
        return chat_completion.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error: {e}")
        return None

# Function to process and write questions and answers to the CSV file
def write_qa_to_csv(qa_output):
    # Open the CSV file in append mode
    with open('champion_qa.csv', mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)

        # Split the output by lines, skip empty lines, and ensure no empty newlines
        lines = qa_output.split('\n')
        clean_lines = [line for line in lines if line.strip()]  # Remove lines with only newlines

        # Write each non-empty line to the CSV file
        for line in clean_lines:
            # Split by the first comma to separate question and answer
            qa_pair = line.split('", "', 1)  # Adjust splitting to match the format of output
            if len(qa_pair) == 2:
                question = qa_pair[0].strip().strip('"')  # Clean up leading/trailing spaces and quotes
                answer = qa_pair[1].strip().strip('"')
                writer.writerow([question, answer])

# Function to read the CSV file and process each champion's background story
def process_champions_from_csv(file_path):
    successful_champions = 0

    with open(file_path, newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)

        for row in reader:
            url_name = row['url-name']
            korean_name = row['korean-name']
            background_story = row['background-story']

            print(f"Processing {korean_name} ({url_name})...")

            qa_output = generate_questions_and_answers(background_story)

            if qa_output:
                # Remove empty lines from qa_output
                qa_output = '\n'.join([line for line in qa_output.split('\n') if line.strip()])
                print(qa_output)

                # Write the Q&A to the CSV file
                write_qa_to_csv(qa_output)
                print(f"Successfully wrote Q&A for {korean_name}")
                successful_champions += 1
            else:
                print(f"Failed to generate questions for {korean_name}.")

    # Print the total number of successfully processed champions
    print(f"\nTotal number of champions successfully processed: {successful_champions}")

# Create an empty CSV file with the column headers "q" and "a"
def create_empty_csv():
    with open('champion_qa.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(["q", "a"])  # Writing header for the CSV

# Path to the CSV file with champion background stories
csv_file_path = 'champion_bs.csv'

# Create the CSV file with headers before processing
create_empty_csv()

# Process the champions from the CSV file
process_champions_from_csv(csv_file_path)
```

```python
import os
import torch
import pandas as pd
from datasets import Dataset, load_from_disk
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, TrainingArguments
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer, setup_chat_format
import bitsandbytes as bnb
from google.colab import userdata
from huggingface_hub import login

login(userdata.get('HF_TOKEN'))
```

```python
BASE_MODEL = "google/gemma-2-2b-it"
DATASET_PATH = "./lol-qa-dataset"
OUTPUT_MODEL_PATH = "./gemma-2-2b-lol"
```

```python
qa_df = pd.read_csv("champion_qa.csv")
qa_df.head()
```

```python
qa_dataset = Dataset.from_pandas(qa_df)
qa_dataset.save_to_disk(DATASET_PATH)
```

```python
# Set up device and dtype
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# if torch.cuda.get_device_capability()[0] >= 8:
#     torch_dtype = torch.bfloat16
#     attn_implementation = "flash_attention_2"
# else:
#     torch_dtype = torch.float16
#     attn_implementation = "eager"
# gemma에선 flash attention사용을 권장하지 않고, eager에 맞춰져 있다고 함.

torch_dtype = torch.float16
attn_implementation = "eager"

print(f"device: {device}\ndtype: {torch_dtype}\nattention implementation: {attn_implementation}")
```

```python
# QLoRA configuration
qlora_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch_dtype,
    bnb_4bit_use_double_quant=True,
)

# Load model with QLoRA
model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    quantization_config=qlora_config,
    device_map="auto",
    attn_implementation=attn_implementation
)
```

```python
# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL, trust_remote_code=True)
```

```python
# Function to generate responses
def generate_response(prompt, model, tokenizer, temperature=0.1):
    inputs = tokenizer(prompt, return_tensors="pt").to(device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=256,
        do_sample=temperature > 0,
        temperature=temperature
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=False)

# Test the model
test_prompt = """<start_of_turn>user
"조이는 아우렐리온 솔한테 어떤 약속을 했어?<end_of_turn>
<start_of_turn>model
"""

response = generate_response(test_prompt, model, tokenizer)
print(response)
```

```python
# Print original model structure
total_layers = 0
for name, module in model.named_modules():
    # print(f"{name}\t\t{type(module)}")
    total_layers += 1
print(f"Total number of layers: {total_layers}")
```

```python
# Print all the linear (that is, lora applicable) layers
total_layers = 0
for name, module in model.named_modules():
    if isinstance(module, bnb.nn.Linear4bit):
        names = name.split('.')
        layer_name = names[-1]
        if layer_name != 'lm_head':
            # print(f"{name}\t\t{type(module)}")
            total_layers += 1
print(f"Total number of layers: {total_layers}")
```

```python
# Return target linear layer module types for LoRA
def find_linear_layers(model):
    linear_layers = set()
    for name, module in model.named_modules():
        if isinstance(module, bnb.nn.Linear4bit):
            names = name.split('.')
            layer_name = names[-1]
            if layer_name != 'lm_head':
                linear_layers.add(layer_name)
    return list(linear_layers)

lora_target_modules = find_linear_layers(model)
lora_target_modules
```

```python
# LoRA configuration
lora_config = LoraConfig(
    r=16, # lora dim
    lora_alpha=32, # W + (alpha / r) * (A dot B)
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=lora_target_modules
)

# Apply LoRA to the model
# model, tokenizer = setup_chat_format(model, tokenizer)
# setup_chat_format()을 사용하면 스페셜 토큰이 (gemma에 맞지 않게) <|im_start|>와 같은 것으로 강제되는 에러가 있음.

model = get_peft_model(model, lora_config)
```

```python
lora_layer_count = 0
lora_a_count = 0
lora_b_count = 0
for name, module in model.named_modules():
    if "lora" in name.lower():
        if isinstance(module, torch.nn.Linear):
            if "lora_A" in name:
                lora_a_count += 1
            elif "lora_B" in name:
                lora_b_count += 1
        # print(f"{name}\t\t{type(module)}")
        lora_layer_count += 1
print(f"Total LoRA-related modules: {lora_layer_count}")
print(f"LoRA A matrices: {lora_a_count}")
print(f"LoRA B matrices: {lora_b_count}")
print(f"Total LoRA rank pairs (A+B): {lora_a_count + lora_b_count}")
```

```python
# Load and preprocess dataset
dataset = load_from_disk(DATASET_PATH)
print(len(dataset))
print(type(dataset))

def format_chat_prompt(example):
    chat_data = [
        # Note: "system" role을 사용하지 않음.
        {"role": "user", "content": example["q"]},
        {"role": "assistant", "content": example["a"]}
    ]
    example["text"] = tokenizer.apply_chat_template(chat_data, tokenize=False)
    return example

dataset = dataset.map(format_chat_prompt, num_proc=4)
print(len(dataset))
print(type(dataset))
```

```python
# Training arguments
training_args = TrainingArguments(
    output_dir=OUTPUT_MODEL_PATH,
    per_device_train_batch_size=1, # steps_per_epoch = ceil(total_samples / (batch_size * gradient_accumulation_steps))
    gradient_accumulation_steps=10, # total_samples means len(dataset)
    num_train_epochs=10,
    learning_rate=2e-4,
    fp16=False,
    bf16=False,
    logging_steps=len(dataset)//10,
    optim="paged_adamw_32bit",
    logging_dir="./logs",
    save_strategy="epoch",
    evaluation_strategy="no",
    do_eval=False,
    group_by_length=True,
    report_to="none"
)

# Initialize trainer
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=lora_config,
    dataset_text_field="text",
    max_seq_length=512,
    tokenizer=tokenizer,
    args=training_args,
    packing=False,
)

# Train the model
trainer.train()

# Save the final model
trainer.model.save_pretrained(OUTPUT_MODEL_PATH)
tokenizer.save_pretrained(OUTPUT_MODEL_PATH)

print(f"Training completed. Model saved to {OUTPUT_MODEL_PATH}")
```

# 3. Merge, Upload, Run Inference

```python
from peft import AutoPeftModelForCausalLM
import torch

new_model = "./gemma-2-2b-lol"

# Load PEFT model on CPU
model = AutoPeftModelForCausalLM.from_pretrained(
    new_model,
    torch_dtype=torch.float16,  # Use float16 to save memory
    low_cpu_mem_usage=True,
    device_map='cpu',  # Explicitly map everything to the CPU
)

# Move all model parameters to the CPU before merging
model = model.to('cpu')  # Force all components to be on the CPU

# Merge LoRA and base model and save
merged_model = model.merge_and_unload()

# Save the merged model with proper serialization
merged_model.save_pretrained("gemma-2-2b-lol-merged", safe_serialization=True, max_shard_size="2GB")
```

```python
## Load merged Model
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, pipeline

OUTPUT_MODEL_PATH = "./gemma-2-2b-lol-merged"
BASE_MODEL_PATH = "google/gemma-2-2b-it"

# Reload tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL_PATH)

model = AutoModelForCausalLM.from_pretrained(
    OUTPUT_MODEL_PATH,
    device_map="auto",
    torch_dtype=torch.float16
)
```

```python
## Upload to HuggingFace (requires HF 'write' token)
HF_TOKEN=userdata.get('HF_WRITE_TOKEN')
REPO_NAME="gemma-2-2b-lol"

merged_model.push_to_hub(
    REPO_NAME,
    use_temp_dir=True,
    use_auth_token=HF_TOKEN
)
tokenizer.push_to_hub(
    REPO_NAME,
    use_temp_dir=True,
    use_auth_token=HF_TOKEN
)
```
